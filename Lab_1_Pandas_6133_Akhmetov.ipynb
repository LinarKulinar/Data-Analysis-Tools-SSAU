Анализируя прошлую таблицу, стало интересно каким элементам можно доверять и я решил подсчитать сколько человек попало в расчёт данных каждой ячейки таблицы

```python
group_data = data[['Happiness in life', 'Healthy eating', 'Eating to survive']]\
  .groupby(['Happiness in life', 'Healthy eating'], as_index=False).count()
# group_data.head(10)
group_data.pivot(index='Happiness in life', columns=['Healthy eating'])
```

Тут уже видно, что в действительности первая строчка нерепрезенертативна т.к. там очень мало людей учавствовало в опросе в каждоя ячейке.
