```python
data[['Happiness in life', 'Healthy eating']]\
    .groupby('Happiness in life')\
    .mean()\
    .sort_values('Healthy eating', ascending=False)\
```

Интересным заключением оказался факт, что чем больше довольность своей жизнью у людей, то выше и уровень заботы о своём здоровье. Ну и соответтсвенно в обрантную сторону: чем лучше человек заботится о своём здоровье, тем выше у него уровень довольства собственной жизнью.

```python
data[['Happiness in life', 'Healthy eating']]\
    .groupby('Healthy eating')\
    .mean()\
    .sort_values('Happiness in life', ascending=False)\
```
